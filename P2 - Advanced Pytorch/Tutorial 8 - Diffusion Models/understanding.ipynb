```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats.distributions import norm
```

```python
x = np.linspace(-10, 10, 100000)
normal = norm()
plt.plot(x, normal.pdf(x))
```

```python
rv1 = norm.rvs(size=100000)*0.5
rv2 = norm.rvs(size=100000)*2 - 5
rv = rv1*rv2**2
_ = plt.hist(rv, bins=1000, density=True)
```
